```python
worksheets = [worksheet1,worksheet2,worksheet3]
# temp = pd.read_excel("test.xlsx",sheet_name="1")
# worksheets = [temp]
```

```python
import smtplib
import os
from email.mime.text import MIMEText
#发送多种类型的邮件
from email.mime.multipart import MIMEMultipart

from email.mime.application import MIMEApplication
from time import sleep

def run_email(email,body,file_path,name):
    msg_from = 'xxxxxxx@qq.com'  # 发送方邮箱
    passwd = 'xxxxxxxxxx'
    
    to= [email] #接受方邮箱
    
    #设置邮件内容
    #MIMEMultipart类可以放任何内容
    msg = MIMEMultipart()
    conntent=body
    #把内容加进去
    msg.attach(MIMEText(conntent,'plain','utf-8'))
    
    #添加附件
    with open(file_path, 'rb') as f:
        attachment = MIMEApplication(f.read(), _subtype=os.path.splitext(file_path)[1][1:])
        attachment.add_header('Content-Disposition', 'attachment', filename=os.path.basename(file_path))
    msg.attach(attachment)   #加入到邮件中
    
    #设置邮件主题
    msg['Subject']="关于保研绩点预计算的通知与核算--NoReply"
    
    #发送方信息
    msg['From']=msg_from
    
    #开始发送
    
    #通过SSL方式发送，服务器地址和端口
    s = smtplib.SMTP_SSL("smtp.qq.com", 465)
    # 登录邮箱
    s.login(msg_from, passwd)
    #开始发送
    s.sendmail(msg_from,to,msg.as_string())
    print(f"{name} 邮件发送成功")
    sleep(10)
```

```python
for worksheet in worksheets:
    for i in range(len(worksheet)):
        row = worksheet.loc[i]
        email = str(row["学号"])+"@whu.edu.cn"
        name = row["姓名"]
        score = round(row["保研绩点"],3)
        major = row["专业"]
        idx = row["idx"]
        file_path = f"personal/{name}.csv"
        hint = row["备注"].split(";") if type(row["备注"])==str else None
        message = f"亲爱的{name}同学:\n\n您好!\n\n根据程序计算，你的保研预排名绩点是{str(score)}，是{major}专业的第{idx}位，您的核算结果见附件。您可以自行进行核对，若与实际不符，请联系杨一鸣同学进行核对。\n本次预排名绩点为非官方排名，意在对同学们的保研规划进行指导。预排名结果与正式保研无关，望悉知。本排名解释权归化学与分子科学学院学生会所有。\n\n"
        appen = f"请注意，系统显示在前几个学期的学习中，保研考查课{hint[0]}门，您{hint[1]}门，这些未修课程并未计入此次计算，但是不修完课程可能导致您不能保研。\n\n" if hint else ""
        end = "祝同学学业顺利，身体健康！\n\n化学与分子科学学院学生会"
        body = message+appen+end
        # print(name)
        run_email(email,body,file_path,name)
        
        
```
